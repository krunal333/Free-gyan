```python
import requests
from pyquery import PyQuery as pq
```

```python
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/?readPreference=primary&appname=MongoDB%20Compass%20Community&ssl=false')
filter={}
result = client['gyan']['urls'].find(
  filter=filter
)
db = client['gyan']
```

```python
import re
def strip_to_num(str):
    try:
        return int(re.sub('\D', '', str))
    except:
        return 0
```

```python
def value_to_int(x):
    if 'k' in x:
        if len(x) > 1:
            return strip_to_num(x) * 1000
        return 1000.0
    if 'm' in x:
        if len(x) > 1:
            return strip_to_num(x) * 1000000
        return 1000000.0
    if 'b' in x:
        return strip_to_num(x) * 1000000000
    return int(x)
```

```python
for item in result:
    start_url = 'https://www.coursera.org/search?query='+item['name'].replace(' ','+')
    r = requests.get(start_url)
    d = pq(r.text)
    pages = int((d('.pagination-controls-container .box.number')[-1]).text)
    
    for x in range (1,pages+1):
        r = requests.get(start_url+'&page='+str(x)+'&index=prod_all_products_term_optimization')
        d = pq(r.text)
        for li in d('.ais-InfiniteHits-list li'):

            course = {}

            course['link'] =  ('https://www.coursera.org' + d(li)('.rc-DesktopSearchCard.anchor-wrapper').attr('href'))
            course['name'] =  (d(li)('.color-primary-text.card-title.headline-1-text').text())
            course['university'] = (d(li)('.partner-name.m-b-1s').text())
            if (d(li)('.ratings-text').text()):
                course['rating'] =   float(d(li)('.ratings-text').text())
            else:
                course['rating'] =  None
            course['rating_count'] =  strip_to_num(d(li)('.ratings-count > span').text())
            if (d(li)('.enrollment-number').text()):
                course['enrolled_students'] =   value_to_int(d(li)('.enrollment-number').text())
            else:
                course['enrolled_students'] =  None
            course['level'] =   (d(li)('.difficulty').text()) 
            course['cover'] =   (d(li)('.image-wrapper img').attr('src'))

            print (x)
            db.course.insert_one(course)
```
